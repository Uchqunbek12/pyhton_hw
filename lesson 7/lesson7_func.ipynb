```python
#1 .
def is_prime (n):
    a = int(n ** 0.5)
    for i in range (2,a+1):
        if n %i == 0:
           return False
    return True

print("4 is ", is_prime(4))
print("7 is ", is_prime(7))
```

```python
#2
def digit_sum(k):
    sum=0
    while k>0:
       sum += k % 10 
       k = k // 10
    return sum

print(digit_sum(24))
print(digit_sum(502))
```

```python
# 3
def power_2 (n,k):
    if n < (2 ** (k+1)):
        return [2**i for i in range(k+1)]
    return power_2(n, k + 1)

n = 10
k = 0
print(power_2(n,k))
```
